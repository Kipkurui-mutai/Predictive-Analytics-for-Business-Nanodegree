```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
diamonds = pd.read_csv('/content/drive/MyDrive/UDACITY/Project 1/diamonds.csv')
new_diamonds = pd.read_csv('/content/drive/MyDrive/UDACITY/Project 1/new-diamonds.csv')
predicted = pd.read_excel('/content/drive/MyDrive/UDACITY/Project 1/predicted.xlsx')
```

```python
x = diamonds['carat'].values.reshape(-1, 1)
y = diamonds['price'].values.reshape(-1, 1)
```

```python
linear_regressor = LinearRegression()
linear_regressor.fit(x, y)
```

```python
y_pred= linear_regressor.predict(x)
```

```python
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)
ax.text(0.95, 0.01, 'KM',verticalalignment='bottom', horizontalalignment='right',transform=ax.transAxes, color='green', fontsize=15)
plt.scatter(x, y)
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title("Price vs Carat")
plt.show()
plt.savefig('/content/drive/MyDrive/UDACITY/Project 1/prices_vs_Carat.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)
ax.text(0.95, 0.01, 'KM',verticalalignment='bottom', horizontalalignment='right',transform=ax.transAxes, color='green', fontsize=15)
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title("Price vs Carat")
plt.show()
plt.savefig('/content/drive/MyDrive/UDACITY/Project 1/LRprices_vs_Carat.png', bbox_inches='tight')
```

```python
x1 = predicted['carat'].values.reshape(-1, 1)
y1 = predicted['Predicted_prices'].values.reshape(-1, 1)
```
